```python
import threading
import time

def worker(num):
    """线程执行函数"""
    print(f'线程 {num} 开始')
    time.sleep(2)  # 模拟耗时操作
    print(f'线程 {num} 结束')

# 创建线程
threads = []
for i in range(10000):
    t = threading.Thread(target=worker, args=(i,))
    threads.append(t)
    t.start()  # 启动线程

# 等待所有线程完成
for t in threads:
    t.join()

print("所有线程执行完毕")
```
